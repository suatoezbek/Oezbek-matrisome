```python
import pandas as pd
import re
```

```python
candidate_list = pd.read_excel("Candidate Liste.xlsx",index_col=0)
candidate_list.index = candidate_list.index.str.replace("Nematostella_vectensis_","")
candidate_list.head()
```

```python
# Assuming limma is your DataFrame
limma = pd.read_csv(r"GideonBergheim_P1482_newDB_V3\data_analysis_results_V3\Limma_results_V3.csv")

# Split the index at '|'
limma.gene_name = limma.gene_name.str.split('|')

# Expand the rows
result_limma = limma.explode('gene_name')

result_limma.set_index("gene_name",inplace=True)
result_limma.head()
```

```python
suppl_limma = result_limma[["protein_id","max.qupm",'logFC','pvalue', 'fdr',"comparison",'hit_annotation', 'AveExpr']]
suppl_limma.columns =  ["protein_id","qupm",'logFC','pvalue', 'fdr',"comparison",'hit_annotation', 'AverageExpression']
suppl_limma.loc[suppl_limma.comparison == "(Larvae - Adult) - (PrimaryPolyp - Adult)", "comparison"] = "PrimaryPolyp - Larvae"
```

```python
front = re.compile(r"NV2t0+")
back = re.compile(r"\d{3}\.\d")

def replace_patterns(input_string):
    output_string = re.sub(front, "NV2.", input_string)
    output_string = re.sub(back, "", output_string)
    return output_string

suppl_limma.index = suppl_limma.index.map(replace_patterns)
```

```python
suppl_limma.head()
```

```python
suppl_limma.to_csv(r"supplementary_data\Full_MS_data.csv")
suppl_limma.to_excel(r"supplementary_data\Full_MS_data.xlsx")
```

```python
candidate_list_excerpt = candidate_list[["gene_names_for_scT ",'short_Name','new_manual_annotation',
       'Division manual', 'Category manual', 'Protein Class', 'Protein Subclass','SignalP prediction']]
candidate_list_excerpt.columns= ["gene_name",'short_Name','description',
       'matrisome_division', 'matrisome_category', 'protein_class', 'protein_subclass','signal_peptide']
#candidate_list_excerpt.head()
candidate_list_excerpt.reset_index(inplace=True)
candidate_list_excerpt=candidate_list_excerpt.set_index("gene_name")
```

```python
candidate_list_excerpt.to_csv(r"supplementary_data\small_candidate_list.csv")
candidate_list_excerpt.to_excel(r"supplementary_data\small_candidate_list.xlsx")
```

```python
matrisome_MS = suppl_limma.loc[suppl_limma.index.isin(candidate_list_excerpt.index)]
```

```python
result = matrisome_MS.merge(candidate_list_excerpt, left_index=True, right_index=True, how='left')
```

```python
indices_to_add = candidate_list_excerpt.index.difference(result.index)
# Select rows from candidate_list_excerpt with those indices
rows_to_add = candidate_list_excerpt.loc[indices_to_add]
# Concatenate the rows
final_result = pd.concat([result, rows_to_add])
```

```python
curated_matrisome_MS_results = final_result[["ProteinID","short_Name","description","matrisome_division","matrisome_category","AverageExpression","qupm","comparison","logFC","pvalue","fdr","hit_annotation"]]
curated_matrisome_MS_results= curated_matrisome_MS_results.sort_values(["matrisome_division","matrisome_category","description"])
curated_matrisome_MS_results = curated_matrisome_MS_results.drop_duplicates()
curated_matrisome_MS_results = curated_matrisome_MS_results.reset_index().set_index(["gene_name","ProteinID","short_Name","description","matrisome_division","matrisome_category","AverageExpression","qupm","comparison"])
```

```python
curated_matrisome_MS_results.to_csv(r"supplementary_data\curated_matrisome_MS_results.csv")
curated_matrisome_MS_results.to_excel(r"supplementary_data\curated_matrisome_MS_results.xlsx", merge_cells=True)
```
